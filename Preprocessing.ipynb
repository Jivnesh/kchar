### Decide vocabulory size

```python
import collections
cnt = collections.Counter()
f = open("./data/san/Sanskrit_Corpus.txt", "r")
lines = f.readlines()
count = 0
for line in lines:
    for word in line.split():
        cnt[word] += 1
```
